# Create function that does all the computations

```python
def stock_compare(stocks,start,end,SMA):
    df=yf.download(stocks,start=start,end=end)
    data=df.Close.to_frame()
    data['return']=np.log(data.Close.div(data.Close.shift(1)))
    data['SMA_S']=data.Close.rolling(int(SMA[0])).mean()
    data['SMA_L']=data.Close.rolling(int(SMA[1])).mean()
    data.dropna(inplace=True)
    data['position']=np.where(data['SMA_S']>data['SMA_L'],1,-1)
  
    data['stragey']=data['return']*data['position'].shift(1)
    data.dropna(inplace=True)
    ret=np.exp(data['stragey'].sum())
    std=data['stragey'].std()*np.sqrt(252)
    
    return ret, std
```

```python
stock_compare("SPY","2000-01-01","2020-01-01",(60,200))
```
